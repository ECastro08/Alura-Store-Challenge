```python
import pandas as pd
import requests
import numpy as np
```

## **Extraccion de los datos**

```python
url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
```

```python
response = requests.get(url)
data = response.json()
df_telecom = pd.DataFrame(data)
df_telecom
```

### **Normalizando los datos**

```python
df_telecom = pd.json_normalize(data)
df_telecom
```

```python
# conviertiendo a formato numerico
df_telecom['account.Charges.Total'] = pd.to_numeric(df_telecom['account.Charges.Total'],errors='coerce')
```

```python
# reviso nuevamente la informacion del dataset
df_telecom.info()
```

```python
#Agrupando un subdataset de los valores nulos detectados en el bloque anterior
df_nulls = df_telecom[df_telecom['account.Charges.Total'].isnull()]
```

```python
df_nulls['customer.tenure'].value_counts()
```
